Prenons les systèmes suivants :

$$G_{BO_0} = \frac{1}{0.5s + 1}$$

$$G_{BO_1} = \frac{1}{s} \cdot \frac{1}{0.5s + 1}$$

$$G_{BO_2} = \frac{1}{s^2} \cdot \frac{1}{0.5s + 1}$$

et observons leurs réponses indicielles, ainsi qu'à une rampe lorsqu'ils sont asservis par un retour unitaire.

> Exécutez les cellules suivantes après avoir bien exécuté la toute première du notebook ou si vous voulez directement obtenir tous les résultats de chaque cellule, vous pouvez utiliser le bouton "Run All".

>**Note :** Afin de faciliter le calcul de la fonction de transfert en boucle fermée, nous allons utiliser la fonction `feedback` qui la détermine à partir de sa fonction en boucle ouverte et de son gain de retour (par défaut : 1).

```python
# Numérateur et dénominateur du système de base
num = 1
den = [0.5, 1]

# Système de classe 0
G_BO_0 = ml.tf(num, den)   # Système en Boucle Ouverte
G_BF_0 = ml.feedback(G_BO_0)  # Système en Boucle Fermée
print(f"G_BO_0 = {G_BO_0}")
print(f"G_BF_0 = {G_BF_0}")

# Système de classe 1
G_BO_1 = ml.tf(1,[1, 0]) * G_BO_0 # On peut multiplier des fonction de transfert et pour intégrer il suffit de multiplier notre système de classe 0 par 1/s => pas de terme indépendant au dénominateur
G_BF_1 = ml.feedback(G_BO_1)
print(f"G_BO_1 = {G_BO_1}")
print(f"G_BF_1 = {G_BF_1}")

# Système de classe 2
G_BO_2 = ml.tf([1, 1],[1, 0]) * G_BO_1 # On peut multiplier des fonction de transfert et pour obtenir un système de classe 3, il faut intégrer 1 fois notre système de classe 1 auquel il faut ajouter un zéro afin d'obtenir un système stable (cf. plus tard dans le cours) => il faut multiplier notre système de classe 1 par (s+1)/s => un zéro aun numérateur et pas de terme indépendant au dénominateur
G_BF_2 = ml.feedback(G_BO_2)
print(f"G_BO_2 = {G_BO_2}")
print(f"G_BF_2 = {G_BF_2}")
```

## Réponses à un échelon (= réponses indicielles)

```python
# Tracé des différentes réponses indicielles
rlf.step(sys_list=[G_BF_0, G_BF_1, G_BF_2], sys_names=["G_BF_0", "G_BF_1", "G_BF_2"], plot_infos=True)
```
